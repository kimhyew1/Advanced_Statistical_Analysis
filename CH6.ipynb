```python
class dog:
    def __init__(self, name, age):
        # self.__name = name
        # self.__age = age   사용자가 직접적으로 변수에 접근할 수 없음
        self.name = name
        self.age = age

    def bark(self):
        print("bark! bark!")

    def get_info(self):
        return f"{self.name} is {self.age} years old."
```

위와 같은 형태를 "캡슐화" 시켰다고 말할 수 있음

```python
dog1 = dog("max", 3)
dog2 = dog("silky", 10)

dog1.bark()
dog2.get_info()

dog2.name   # 변수에 직접적으로 접근할 때에는 괄호 필요 없음
```
